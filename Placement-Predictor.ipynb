```python
import numpy as np
import pandas as pd




df=pd.read_csv('/content/placement.csv')
df.head()
```

```python
df.shape
```

```python
df = df.iloc[:, 1:]
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['cgpa'], df['iq'], c=df['placement'])
```

```python
input = df.iloc[:, 0:2]
input
```

```python
output = df.iloc[:, -1]
output
```

```python
from sklearn.model_selection import train_test_split
i_train, i_test, o_train, o_test = train_test_split(input, output, test_size=0.1)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(i_train, o_train)
```

```python
o_predict = clf.predict(i_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(o_test, o_predict)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(i_train.values, o_train.values, clf=clf, legend=2)
```
