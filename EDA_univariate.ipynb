#**EDA using univariate analysis**

```python
import pandas as pd
import seaborn as sns
```

```python

df = pd.read_csv('train.csv')

df.head()
```

#**1. Categorical Data**
eg. Survived, Pclass, (Name), Sex, Embarked

##a. Countplot

```python
#Survived column has categorical data (either 0 or 1) but the data type is
# actually numeric(0,1). So we cannot use-> "sns.countplot(df['Survived'])"


#sns.countplot(x='Pclass', data=df)
sns.countplot(x='Survived', data=df)
df['Survived'].value_counts()
```

##b. Piechart

```python
df['Pclass'].value_counts().plot(kind='pie', autopct='%.2f')
df['Pclass'].value_counts()
```


#**2. Numerical Data**

eg. Age, Fare



##a. Histogram


```python
import matplotlib.pyplot as plt


plt.hist(df['Age'])
plt.xlabel("Age")
plt.ylabel("Frequency")
```

##b. Distplot

```python
sns.distplot(df['Age'])

#x-axis: Age
#y-axis: Probability of being that age(*100)--PDF(Prob. Density Function)

#Distplot shows both histogram and KDE(Kernel Density Function)
```

##c. Boxplot
gives 5number summary->
1.Minimum value (except outliers)
     2.Q1 (25th percentile)
     3.Median (Q2 / 50th percentile)
     4.Q3 (75th percentile)
     5.Maximum value (except outliers)

Any value outside minimum = Q1-1.5*IQR and maximum = Q3+1.5*IQR  are called potential outliers of the dataset

```python
sns.boxplot(df['Age'])
```
