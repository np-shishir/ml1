```python
import seaborn as sns
import pandas as pd
```

```python
#seaborn provides multiple in-built data-sets
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins['species'].value_counts()
```

```python
penguins['island'].value_counts()
```

```python
sns.scatterplot(data = penguins, x="flipper_length_mm", y='body_mass_g', hue='species')
```

```python
sns.stripplot(data=penguins, y="body_mass_g", x="species", hue="island", dodge=True)
```

```python
sns.histplot(data=penguins, x="body_mass_g", bins=15, hue="sex")
```

```python
#regression plot
sns.regplot(data=penguins, x="body_mass_g", y="flipper_length_mm")
```

```python
sns.lineplot(data=penguins, x="body_mass_g", y= "flipper_length_mm")
```

```python
sns.barplot(data=penguins, x="species", y= "body_mass_g", hue="sex")
```

```python
sns.boxplot(data=penguins, x="species", y="body_mass_g", hue="sex")
```

```python
sns.kdeplot(data=penguins, x="body_mass_g", hue="sex")
```
